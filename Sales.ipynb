```python
#IMS-Selling to customer(Sales.ipynb)

import json
import time
record = []   

fd = open('product_data.json', 'r')
json_data = fd.read()
prod_data = json.loads(json_data)

bold = '\33[1m'
print(bold + '--------------Welcome to IMS-----------------')
print('20% discount on purchase more than 2000 Rs :)')
print('35% discount on purchase more than 6000 Rs :)')
name = str(input('Enter your name --> '))
print('Welcome ',name)  
pID = str(input('Enter product ID --> '))

if pID in prod_data:                                                    
    
    print('Rescective product is -->',prod_data[pID]['name'],'& price -->',prod_data[pID]['value'])   
    quant = int(input('Enter Quantity --> '))
           
    if quant > prod_data[pID]['quantity']:                              
        
        print("Sorry,we don't have",quant,"items currently in stock!")
        
    else:
        
        print('*****************')
        print('product ID:',pID)
        print('name:',prod_data[pID]['name'])
        print('brand:',prod_data[pID]['brand'])
        print('rating:',prod_data[pID]['rating'])
        print('Price:',prod_data[pID]['value'])
        print('Manufacturing:',prod_data[pID]['man_date'])
        print('Exp_date:',prod_data[pID]['exp_date'])
        print('*****************')
        
        print('------------Billing-------------')
        print('Wait,your bill is on the way!!')
        time.sleep(6)
        print(name,'you have ordered: ',quant, 'items')
        print('Total payable amount: ',int(quant) * int(prod_data[pID]['value']),"Rs")
        print('Transaction details: ',time.ctime())                       
        print('--------------------------------')

        if  2000 < int(quant) * int(prod_data[pID]['value']) < 6000:      
            
            print('Checking for any discount.....')
            time.sleep(5)
            print('---> Hey,you got 20% discount, Pls pay only: ',(int(quant) * int(prod_data[pID]['value']) - int(20/100 * int(quant) * int(prod_data[pID]['value']) )),'Rs')
                
        elif int(quant) * int(prod_data[pID]['value']) > 6000:              
        
            print('Checking for any discount.....')
            time.sleep(5)
            print('---> Hey,you got flat 35% discount, Pls pay only: ',(int(quant) * int(prod_data[pID]['value']) - int(35/100 * int(quant) * int(prod_data[pID]['value']) )),'Rs')
    
        else:
            print('Checking for any discount.....')
            time.sleep(5)
            print('---> Not eligible for discount!')
            
        if quant < prod_data[pID]['quantity']:
            
         
            prod_data[pID]['quantity'] = prod_data[pID]['quantity'] - quant        
    
            record.append(str(pID) + ', [' + name + '], [Quantity left: ' + str(prod_data[pID]['quantity']) + '], [' + prod_data[pID]['name'] + '], [Quant Ordered:'+ str(quant) + '], [' + time.ctime() + '], [Total bill:' + str(int(quant) * int(prod_data[pID]['value'])) +'], [If 20% Discount:' + str((int(quant) * int(prod_data[pID]['value']) - int(20/100 * int(quant) * int(prod_data[pID]['value']) ))) + '], [Or if 35% discount:' + str((int(quant) * int(prod_data[pID]['value']) - int(35/100 * int(quant) * int(prod_data[pID]['value']) ))) + "]\n") 
            record[-1] = record[-1][:-1] + "\n"
        
        print('Thanks for choosing us, come again!') 
    
else:                                                                          

    print(name,'enter correct product ID!') 
    
                                                       
fd = open('sales_rec.txt', 'a')             
for i in record:   
    fd.write(i)         
fd.close()

fd = open('product_data.json', 'w')

js = json.dumps(prod_data)
fd = fd.write(js)
```
