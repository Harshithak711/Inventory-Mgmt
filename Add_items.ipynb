```python
#IMS-Add_items.ipynb(Adding Items or quantity to Inventory)

import json
import time
records = []
q_changed = []

fd = open('product_data.json', 'r')
json_data = fd.read()
prod_data = json.loads(json_data)

bold = '\33[1m'
print(bold +'---------Welcome Admin--------')
u_choice = int(input(bold + '(For adding new product press 1) or (for adding quantity of existing product,press 0) -->'))


if u_choice == 1:
    
    print('We are ready to add new item :')
    ID = str(input('Enter pID -->'))
    
    if ID in prod_data:
        print('This ID already present in inventory!')
        
    else:
            Name = str(input('Enter name -->'))
            Brand = str(input('Enter brand_name -->'))
            Rating = str(input('Enter rating -->'))
            Price = int(input('Enter price -->'))
            manufacturing = str(input('m_date -->'))
            expiry = str(input('exp_date -->'))
            quantity = int(input('Qauntity -->'))
            print('It require few seconds for addition')
            time.sleep(4)
            print('--> New product successfully added.')
    
            prod_data[ID] = {"name": Name, "brand": Brand, "value": Price, "man_date": manufacturing, "exp_date": expiry, "rating": Rating, "quantity": quantity }
            
            records.append(str(ID) + ', [' + Name + '], [' + str(Brand) + '], [' + Rating + '], [Price:'+ str(Price) + '], [' + manufacturing + '], ['+ expiry + '],[Quantity:' + str(quantity) + '], ['+ time.ctime() + "]\n" ) 
            print('Inventory Updated at [',time.ctime(),']') 
        
    
elif u_choice == 0:
     
    ID = str(input('Enter pID of product for which you want to change quantity -->'))
    
    if ID not in prod_data:
        print('The product for given ID not present in inventory.')
        
    else:
        
        print('--> Respective product is',prod_data[ID]['name'],'and whose current stock is',prod_data[ID]['quantity'],'in inventory.')
        quantity = int(input('Enter additional quantity -->'))
        prod_data[ID]['quantity'] = prod_data[ID]['quantity'] + quantity
        print('It require few seconds')
        time.sleep(4)
        print('Total quantity becomes -->',prod_data[ID]['quantity'])
        print('--> Successfully added',quantity,'more items.')
        print('Inventory Updated at [',time.ctime(),']')
    
        q_changed.append(str(ID) + ', [' + prod_data[ID]['name'] + '], [Added quantity:' + str(quantity) + '], [Total quantity:' + str(prod_data[ID]['quantity']) + '], [' + time.ctime() + "]\n")
        
else :
    
    print('Pls, provide correct response!')
    
fd = open('changed_quantity_record.txt', 'a')             
for i in q_changed:   
    fd.write(i)         
fd.close() 

fd = open('Add_item_record.txt', 'a')             
for i in records:   
    fd.write(i)         
fd.close() 
    
fd = open('product_data.json', 'w')

js = json.dumps(prod_data)
fd = fd.write(js)
```
